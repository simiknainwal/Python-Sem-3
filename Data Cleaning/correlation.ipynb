# Correlation
It is a statistical measure that describes the degree to which two variables are related.

Two variables must be either positively or negatively correlated.

Positive correlation=> directly proportional. +I:Perfect positive correlation.

Negative correlation=> inversely proportional. -I:Perfect negative correlation

0:No correlation.
