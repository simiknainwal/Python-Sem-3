# Z-score
It is an important concept in statistics. It is also called standard score.

It helps to understand if a data value is greater or smaller than mean and how far away it is from mean.

It tells how many standard deviations away a data is from mean.

# Algorithm
1) Read the data.

2) Calculate mean value of data.

3) Calculate standard deviation of data.

4) for i in data

   Calculate Z-score=(i-mean)/standard deviation
   
   if Z-score>3 OR Z-score<-3

   Add to outlier list.

5) Print outliers.

```python
import pandas as pd
df=pd.read_csv("heights-outlier_1.csv")
```

```python
df.columns
```

```python
mean=df.height.mean()
mean
```

```python
std=df.height.std()
std
```

```python

df["z_score"]=(df.height-mean)/std
df.head(100)
```

```python
outliers=df[(df.z_score >3) | (df.z_score <-3)]
outliers
```
