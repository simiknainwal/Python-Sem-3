# Interquartile range

It is used to measure variablity by dividing the dataset into quartiles.

Q1,Q2,Q3 are first,second and third quartiles.

# Algorithm

1) Calculate first Q1 and third Q3 quartile.

2) IQR= (Q3-Q1)

3) Find lower bound.

   Lower bound= Q1-(1.5*IQR)

4) Find upper bound

   Upper bound= Q3+(1.5*IQR)

Anything that lies below lower and above upper bound is an outlier.   

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
df=pd.read_csv("heights-outlier_1.csv")
df.shape
df.head()

```

```python
sns.boxplot(data=df, y="height")
plt.show()
```

```python
df.describe()
```

```python
Q1=df.height.quantile(0.25) #method to calculate percentile.
Q1
```

```python
Q3=df.height.quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
LB= Q1-(1.5*IQR)
LB
```

```python
UB=Q3+(1.5*IQR)
UB
```

```python
df[df.height<LB]
```

```python
df[(df.height<LB) | (df.height>UB)]
```

```python
new_df=df[(df.height>LB) & (df.height<UB)]
# new_df
```

```python
df.height > LB # return type is boolean so 
```

```python
new_df.shape
```

```python
new_df.describe()
```

```python
sns.boxplot(data=new_df, y="height")
plt.show()
```
