```python
import matplotlib.pyplot as plt
```

```python
names=["Alice", "Amit","Simik","Sarbesh"]
sgpa=[9.2,8,7,5]
# plt.bar(names,sgpa,width=0.5,color=["r","g","b","pink"])# (x-axis,y-axis)
plt.bar(names,sgpa,width=0.2,color="skyblue",edgecolor="blue", linewidth=0.5,hatch="*",align="edge",label="SGPA")# (x-axis,y-axis)
plt.legend()
plt.show()
```

x=> The x-axis values

height=> The height of each bar(y-axis values)

width=> The width of each bar (default=0.8)

color=>Color of the bars.

edgecolor=> Color of the bar borders.

linewidth=> Thickness of bar borders.

tick_label=> labels to place under each bar.

align=> How bars are aligned to their x- coordinates.

alpha=>Transparency of bars.

label=> label for dataset

hatch=> adds patterns to bars.

zorder=> controls layer order, high zorder= drawn on top.

```python
import numpy as np
names=["Alice", "Amit","Simik","Sarbesh"]
sgpa=[9.2,8,7,10]
cgpa=[8,4,2,5]
plt.bar(names,sgpa)
plt.bar(names,cgpa)
plt.show()
```

```python
x=np.arange(len(names))
width=0.35
import numpy as np
names=["Alice", "Amit","Simik","Sarbesh"]
sgpa=[9.2,4,7,10]
cgpa=[8,7,10,5]
plt.bar(x-width/2,sgpa,width=width,color="blue",hatch="/",edgecolor="green",label="SGPA")
plt.bar(x+width/2,cgpa,width=width,color="orange",hatch="/",edgecolor="green",label="CGPA")
plt.xticks(x,names)
plt.xlabel("Names")
plt.ylabel("CGPA and SGPA")
plt.legend()
plt.show()
```

```python
import pandas as pd 
df=pd.read_csv("students_sgpa_cgpa.csv")
names=df["Name"]
cgpa=df["CGPA"]
sgpa=df["SGPA"]
import numpy as np
x=np.arange(len(names))
width=0.35
# import numpy as np
plt.bar(x-width/2,sgpa,width=width,color="blue",hatch="/",edgecolor="green",label="SGPA")
plt.bar(x+width/2,cgpa,width=width,color="orange",hatch="/",edgecolor="green",label="CGPA")
plt.xticks(x,names,rotation=45)
plt.xlabel("Names")
plt.ylabel("CGPA and SGPA")
plt.legend()
plt.show()
```
