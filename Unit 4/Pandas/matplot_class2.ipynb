plot()=> Syntax: plt.plot(x,y,**kwargs)

x=>data for x-axis.

**kwargs=>additional customization content.

Markers

'o'=> Shape=> circle.

```python
import matplotlib.pyplot as plt
```

```python
overs=[2,4,6,12,14,16,18,20]
ind=[18,24,32,40,56,74,89,103]
aus=[16,28,36,56,66,70,85,95]

plt.plot(overs,ind,marker='o',ls='--',lw='1',mfc="green",label="IND")
plt.plot(overs,aus,marker='^',ls=':',lw='1',mfc="blue",label="AUS")
plt.legend(title="TEAM",shadow=True)
plt.xlabel("Overs")
plt.ylabel('team')
plt.title("T20 Match IND vs AUS")
plt.show()
```
