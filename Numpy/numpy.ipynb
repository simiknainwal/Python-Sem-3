Array operations

```python
x=np.array([1,2,3])
y=np.array([10,20,30])

print(x+y)
print(x*y)
print(x**2)
```

Broadcast

```python
M=np.array([[1,2,3],[4,5,6]])
V=np.array([10,20,30])
print(M+V) #values of V are added in both rows of M.
```

Scalar Broadcasting

```python
print(2*M)
```

```python
x=np.array([1,16,9])
print(np.sqrt(x))
print(np.exp(x))
```

Statistical Operations

```python
A=np.array([[1,2,3],[4,5,6]],dtype=float)

print(A.sum())
print(A.mean())
print(A.var(),A.std())
print(A.min(),A.max())
```

```python
A=np.array([[1,2,3],[4,5,6]],dtype=float)
print(A.sum(axis=0)) #colomn-wise sum.
```

```python
print(A.sum(axis=1)) #row-wise sum
```

```python
print(np.median(A))
print(np.percentile(A,75))
```

Slicing

```python
D=np.arange(1,13).reshape(3,4)
print(D)
```

```python
print(D[0,0])
print(D[0,:])
print(D[:,0])
print(D[:,1])
print(D[0:2,1:3]) # row,column => only 0,1 rows are included 
```

Linear Algebra Operations

```python
# Dot operator of vectors.
u=np.array([1,2,3])
v=np.array([4,5,6])
print(u@v)
```

```python
# Solve Mx=b
M=np.array([[3,1],[1,2]])
b=np.array([9,8])
x=np.linalg.solve(M,b)
print(x)
```

```python
# Inverse and determinant
Minv=np.linalg.inv(M)
detM=np.linalg.det(M)
print(Minv)
print(detM)
```

```python
# Eigen Values and eigen vectors.
eigvals, eigvecs=np.linalg.eig(M)
print(eigvals) # array of lambda
print(eigvecs) # columns are eigenvectors.
```
